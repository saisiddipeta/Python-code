```python
def outlier_iqr(input_data):
    q1=input_data.quantile(0.25)
    q3=input_data.quantile(0.75)
    iqr=q3-q1
    lower_side = q1-(1.5*iqr)
    upper_side = q3+(1.5*iqr)
    outlier_list = []
    for i in input_data:
        if i>upper_side or i<lower_side:
            outlier_list.append(i)
    print ('outliers are', len(outlier_list))
    return ('lower_side value :', lower_side, 'upper_side value', upper_side)
```

```python
outlier_iqr(df['Amount'])
```

# Data Splitting

```python
x_data = df.drop('Class', axis = 1)
x_data.head()
```

```python
y_data = df['Class']
y_data
```

### As the data is highly imbalanced we need to adjust the data by using Under Sampling and Over Sampling method.  Here we use Under Sampling for this data set.

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=0)
x_resampled, y_resampled = rus.fit_resample(x_data, y_data)
```

```python
x_resampled.shape, y_resampled.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test= train_test_split(x_resampled, y_resampled,test_size=0.2, random_state=25)
x_train
```

```python
y_test
```
